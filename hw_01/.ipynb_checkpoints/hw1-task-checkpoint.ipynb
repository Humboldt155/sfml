```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
def limit_bal_to_char(limit_bal):
    if limit_bal <= 10000:
        return 'A'
    elif 10000 < limit_bal <= 100000:
        return 'B'
    elif 100000 < limit_bal <= 200000:
        return 'C'
    elif 200000 < limit_bal <= 400000:
        return 'D'
    elif 400000 < limit_bal <= 700000:
        return 'E'
    elif limit_bal > 700000:
        return 'F'
    
df['NEW'] = df['LIMIT_BAL'].map(limit_bal_to_char)
df['NEW']
```

```python
# (10) построить большой график (подсказка - используя seaborn) для построения завимисости всех возможных пар параметров
# разным цветом выделить разные значение "default payment next month"
# (но так как столбцов много - картинка может получиться "монструозной")
# (поэкспериментируйте над тем как построить подобное сравнение параметров)
# (подсказка - ответ может состоять из несколькольких графиков)
# (если не выйдет - программа минимум - построить один график со всеми параметрами)
import seaborn

#TODO
# Удаляем ID, он не несет никакой полезной информации
df_numeric = df.drop(['ID'], axis=1)

# Получаем список параметров
all_params = list(df_numeric.columns)

# Разделяем параметры на группы по четыре штуки (будем строить отчеты 4x4)
all_params = list(all_params[x:x+4] for x in range(0, len(all_params), 4))

# Создаем копию листа, будем уменьшать его в цикле, чтобы исключить повторения
reducing_params = all_params

# Запускаем цикл
# Получаем все возможные комбинации
# в каждом цикле получается:
# 1 таблица: Все комбинации первого набора из 4 параметров
# n таблиц (n = количество наборов в reducing_params): Комбинация первого набора со всеми наборами из reducing_params
# Прогодим гарантированно по всем парам, причем ни одна пара не дублируется
# Общее время выполнения - около 3 минут

for param in all_params:
    # создаем сетку 4х4, в которой участвуют параметры из первого набора
    sns.pairplot(df_numeric,
                     vars = param,
                     hue='default.payment.next.month')
    
    # Удаляем параметры первого набора, чтобы не было дублирования, и повтора в следующем цикле
    reducing_params = [p for p in reducing_params if p != param]
    if len(reducing_params) == 0:
        break
    for r_param in reducing_params:
        print('Подготовка параметров: {}'.format(param + r_param))
        # Используем параметры x_vars и y_vars, для того чтобы не было дублирования параметров
        
        sns.pairplot(df_numeric,
                     x_vars = param,
                     y_vars = r_param,
                     hue='default.payment.next.month')
```
